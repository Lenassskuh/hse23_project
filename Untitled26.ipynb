```python
#импортируем библиотеки
import pandas as pd
import math
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Прочитаем заранее созданную таблицу со значениями e-value
df = pd.read_csv('kdm7a.csv', sep=';')
df
```

```python
#Убедимся, что все типы данных верные
df.info()
```

```python
df.index = ['H4', 'H3', 'H2B', 'H2A', 'KDM7A']
df
```

```python
#логарифмируем все значения
log_df = df.applymap(lambda x: -1 * np.log10(x) if x > 0 else 100)
log_df
```

```python
sns.heatmap(log_df, cmap='YlOrRd')
plt.ylabel('Protein')
```
